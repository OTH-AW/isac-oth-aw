#### Beschreibung:
Zeigt die Daten des derzeitig in der Konfiguration gesetzten Projektes als Chart an.

#### Workflow:
Echtdaten werden über die Notebooks data_force_sensor/clean_data aus der MongoDB (Datenbank) ausgelesen. Anschließend werden sie in diesem Notebook visualisiert.

# Daten aufbereiten / visualisieren

```python
%run ../Setup.ipynb

import matplotlib.pyplot as plt
from data_force_sensor import clean_data as data
```

## Plotten

```python
dataset = data.df[data.focusColumns]

fig, axes = plt.subplots(nrows=3, ncols=1)
fig.tight_layout()

c1 = (242/255,148/255,0)
c2 = (98/255,144/255,161/255)
c3 = (121/255,68/255,121/255)

# Fsum
dataset.plot(figsize=(15,12), ax=axes[0], x='time',  y='fsum', color=c1); 
axes[0].set_title('Kraft Summe');
axes[0].set_ylabel("Kraft in N");

# Fx
dataset.plot(ax=axes[1], x='time',  y='fx', color=c2); 
axes[1].set_title('Gemessene Kraft in X-Richtung');
axes[1].set_ylabel("Kraft in N");

# Fy
dataset.plot(ax=axes[2], x='time',  y='fy', color=c3); 
axes[2].set_title('Gemessene Kraft in Y-Richtung');
axes[2].set_ylabel("Kraft in N");
```

#### Alle in einem

```python
fig, axes = plt.subplots(figsize=(50,12))

dataset.plot(color=c1, ax=axes, x='time',  y='fsum'); 
dataset.plot(color=c2, ax=axes, x='time',  y='fx'); 
dataset.plot(color=c3, ax=axes, x='time',  y='fy'); 
axes.set_title('Kraft Summe');
axes.set_ylabel("Kraft in N");
```
