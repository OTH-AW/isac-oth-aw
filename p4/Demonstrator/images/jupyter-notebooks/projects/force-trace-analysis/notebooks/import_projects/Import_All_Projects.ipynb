#### Beschreibung:
Importiert alle Projekte, die im Ordner /data/projects/ abgelegt wurden.

#### Workflow:
Echtdaten werden über die Notebooks data_force_sensor/import_data_to_database bzw. data_machine/import_data_to_database in die MongoDB (Datenbank) geschrieben. Dazu wird zuerst der aktuelle Projektname anhand des Namens des derzeitig zu importierenden Projektnamens in der Konfiguration gesetzt. Anschließend wird der Import gestartet.

```python
# Notebook neu laden
def modulereload(modulename):
    import importlib
    if modulename not in sys.modules:
        importlib.import_module(modulename)
    else:
        importlib.reload(sys.modules[modulename])

# Projektdaten setzen
def setProject(project_path):
    configuration.project = project_path
    configuration.project_name = os.path.basename(project_path).replace(" ", "_")
    configuration.force_sensor_importFileMixedContent = project_path + "/force_sensor/FORCE_KONV_1_Nut2.txt"
    configuration.force_sensor_importFile = project_path + "/force_sensor/FORCE_KONV_1_Nut2_CONVERTED.txt"
    configuration.machine_importFile = project_path + "/machine/TRACE_KONV_1_Nut2_roh.csv"

directory = "../../data/projects"
projects = [f.path.replace('\\', '/') for f in os.scandir(directory) if f.is_dir()]
for project in projects:
    print("##############################")
    print("Project: " + project)
    setProject(project)
    print("Import Force Sensor Data")
    modulereload("data_force_sensor.import_data_to_database")
    print("Import Machine Data")
    modulereload("data_machine.import_data_to_database")
    print("##############################")
```
