#### Workflow:
Echtdaten werden über die Notebooks data_machine/clean_data aus der MongoDB (Datenbank) ausgelesen. Anschließend werden sie in diesem Notebook visualisiert.

# Daten aufbereiten / visualisieren

```python
%run ../Setup.ipynb
%matplotlib inline

import matplotlib.pyplot as plt
from data_machine import clean_data as data
```

## Plotten

```python
dataset = data.df[data.focusColumns]

fig, axes = plt.subplots(nrows=4, ncols=1)
fig.tight_layout()

c1 = (242/255,148/255,0)
c2 = (98/255,144/255,161/255)
c3 = (121/255,68/255,121/255)
c4 = (95/255,140/255,97/255)

# Virtuelle Kraft
dataset.plot(figsize=(15,12), ax=axes[0], x='time',  y='Virtuelle Kraft', color=c1); 
axes[0].set_title('Virtuelle Kraft');
axes[0].set_ylabel("Kraft in N");

# Antriebsmomentensollwert 
dataset.plot(ax=axes[1], x='time',  y='torque_set_f11', color=c2); 
axes[1].set_title('Antriebsmomentensollwert in N');
axes[1].set_ylabel("Kraft in N");

# Antriebswirkleistung
dataset.plot(ax=axes[2], x='time',  y='spindle_power_f9', color=c3); 
axes[2].set_title('Antriebswirkleistung Hauptspindel in W');
axes[2].set_ylabel("Leistung in W");

# Spindeldrehzahl, Istwert in RPM
dataset.plot(ax=axes[3], x='time',  y='spindle_speed_actual_f3', color=c4); 
axes[3].set_title('Spindeldrehzahl, Istwert in RPM');
axes[3].set_ylabel("RPM");
```
